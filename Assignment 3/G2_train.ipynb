# Implementation of Optimus Prime (Transformer)


```python
# List of imports 
from DeepPurpose import utils
from DeepPurpose import dataset
import DeepPurpose.DTI as models
import matplotlib.pyplot as plt
import pandas as pd
import pickle as pickle_rick
import numpy as np
```

##Load data 
We are using both DAVIS and Binding_BD datasets to promote generalization and prevent overfitting.  

This part of the code will download both DAVIS and Binding_BD. The data is split into drugs, targets, binding and binding affinity. The binding affinity is coverted into a logspace and the value for binding affinity is repensent by float values.


```python
DAVIS_X_drugs, DAVIS_X_targets, DAVIS_y = dataset.load_process_DAVIS(path = './data', binary = False, convert_to_log = True, threshold = 30)
print('Drug 1: ' + DAVIS_X_drugs[0])
print('Target 1: ' + DAVIS_X_targets[0])
print('Score 1: ' + str(DAVIS_y[0]))

data_path = dataset.download_BindingDB('./data/BindingDB_All.tsv')
BDB_X_drugs, BDB_X_targets, BDB_y = dataset.process_BindingDB(path = data_path, df = None, y = 'Kd', binary = False, convert_to_log = True, threshold = 30)
print('Drug 1: ' + BDB_X_drugs[0])
print('Target 1: ' + BDB_X_targets[0])
print('Score 1: ' + str(BDB_y[0]))
```

##Combining both data
After the data is download and split into drugs, targets, and binding affinity the both dataset are combined together. 

```python
Comb_drugs = np.concatenate((DAVIS_X_drugs,BDB_X_drugs),axis=None)
Comb_targets = np.concatenate((DAVIS_X_targets,BDB_X_targets),axis=None)
Comb_y = np.concatenate((DAVIS_y,BDB_y),axis=None)
```

We decided to go with a CNN Transformer model and compare it with the DeepDTA solution. Transformers models are similar to LSTMs and RNNs as they are designed to work with sequential data for example natural languages, in order to perform tasks such as translation or text summarization. The main difference between Transformers and RNNs or LSTMs is that the sequential data does not need to be processed in order, this allows for much more parallelization.  


The drugs, targets, and binding affinity is further split into train, validation, test dataset using cold protein. A 70 10 20 split was used to ensure that the  model trained with the majority of the data and the hold out set had enough data validation and test the model to determine how well the model does.

```python
drug_encoding, target_encoding = 'CNN', 'Transformer'
train, val, test = utils.data_process(Comb_drugs, Comb_targets, Comb_y, 
                                drug_encoding, target_encoding, 
                                split_method='cold_protein',frac=[0.7,0.1,0.2],
                                random_seed = 1)
```

For the CNN part of the model, Based on the DeepDTA model, there were three convolutional layers with increasing number of filters. The second layer had double and the third convolutional layer had triple the number of filters in the first one. Therefore through experimentation it was determined the best filter set up was for the CNN block was [32,64,96], this matched the DeepDTA setup. The kernel parameters were tuned through experimentation  it was determined  that the following setup yielded the best results [4,8,12], this matched the DeepDTA setup. A dropout layer of 0.1 was added as well to promote model Generalization. Dropout layers are regularization techniques that are used to reduce the affects of over-fitting by setting the activation of some of the neurons to 0. The transformer model parameters were chosen form the following paper Proteomics Analysis of FLT3-ITD Mutation in Acute Myeloid Leukemia Using Deep Learning Neural Network. In paper it was determined to use mpnn depth of 3, this value was tuned through brute force search. This method of tuning  applies to the other hypermeters for the transformer model as well.

```python
config = utils.generate_config(drug_encoding = drug_encoding, 
                         target_encoding = target_encoding,
                         cls_hidden_dims = [1024,1024,512], 
                         train_epoch = 100, 
                         test_every_X_epoch = 10, 
                         LR = 0.001, 
                         batch_size = 128,
                         hidden_dim_drug = 128,
                         mpnn_hidden_size = 128,
                         mpnn_depth = 3,
                         transformer_dropout_rate = 0.1, 
                         cnn_target_filters = [32,64,96],
                         cnn_target_kernels = [4,8,12] 
                        )
model = models.model_initialize(**config)
model.train(train, val, test)
```

```python
model.save_model('./model_CNN_Transformer_Assignment_3_e100_test')
```

**Evaluations and Experiments**

```python
filename='G2_model.pkl'
Test_model = pickle_rick.load(open(filename, "rb"))

test['predicted'] = Test_model.predict(test)

test['MSE'] = ((test['Label'] - test['predicted']))**2
MSE = (1/len(test['MSE'])) * (test['MSE'].sum())
print('The MSE for the CNN Transformer model is ' + str(MSE))
RMSE = MSE**(1/2)
print('The RMSE for the CNN Transformer model is ' + str(RMSE))
```

```python
filename='deepDTA.pkl'
Test_model = pickle_rick.load(open(filename, "rb"))

test['predicted'] = Test_model.predict(test)

test['MSE'] = ((test['Label'] - test['predicted']))**2
MSE = (1/len(test['MSE'])) * (test['MSE'].sum())
print('The MSE for the deepDTA model is ' + str(MSE))
RMSE = MSE**(1/2)
print('The RMSE for the deepDTA model is ' + str(RMSE))
```

Based on the RMSE values it shows that the deepDTA model was in fact more accurate than the proposed CNN transformer model, one possible reason for this can be that the parameters for the proposed model were not tuned properly. The main limitation was that it takes around 7 hours to train a model, therefore due to time constraints limited tuning was done.
